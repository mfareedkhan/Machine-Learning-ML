# **Machine Learning**

#### **What is Machine Learning?**
> Machine learning is a type of artificial intelligence (AI) that provides computers with the ability to learn without being explicitly programmed. Machine learning focuses on the development of computer programs that can change when exposed to new data.



- As the name suggests,ML allows machines to learn and make decisions smartly. 
- In ML, machines can learn from the data provided or their own experience. It depends upon the type of machine learning.
- Machine study the data (structural / rectangular) patterns and predict the unknown values on behalf of the pattern and make models (equations/algorithms)

#### **How Machine Learning Works?**



#### **Types of Machine Learning**
1. Supervised Learning
2. Unsupervised Learning / Clustering
3. Semi-Supervised Learning
4. Reinforcement Learning

#### **1. Supervised Learning**
`Task driven (predict next value)`
> Supervised learning is the types of machine learning in which machines are trained using well "labelled" training data, and on basis of that data, machines predict the output. The labelled data means some input data is already tagged with the correct output.
- Works under supervision
- Teacher teaches
- Prediction 
- Outcome 

**How Supervised Learning Works?**
- In this type, the algorithm is trained on a labeled dataset, where the input data is paired with the corresponding correct output. 
- The algorithm learns to map inputs to outputs, and once trained, it can make predictions on new, unseen data.
- Supervised learning is a category of machine learning that uses labeled datasets to train algorithms to predict outcomes and recognize patterns.



#### **Types of Supervised Learning:**

| Type | Categories | Uses | Algorithms / Models |
|---|---|---|---|
| Classification (defined labels) | For categories | Diagnostics, Customer Retention, Image Classification, Identity Fraud Detection | Naive Bayes, Support Vector Machine (SVM), Random Decision Forest, Discriminant analysis, K-Nearest Neighbours (K-NN), Neural networks |
| Regression (no labels defined) | For numerical data | Advertising popularity prediction, weather forecasting, market forecasting, estimating life expectancy, population growth prediction | Linear Regression, Logistic Regression, Support Vector Regression (SVR), Gaussian Process Regression (GPR), Ensemble Methods, Decision trees |

#### **2. Unsupervised Learning**
`Data driven (find hidden patterns)`
> Unsupervised learning is the type of machine learning in which machines are trained using "unlabelled" training data. The algorithm explores the data and draws inferences from datasets to describe hidden structures from "unlabelled" data.
- No supervision
- No teacher
- Self learning
- No labeling of data
- Find patterns by itself

**How Unsupervised Learning Works?**



#### **Types of Unsupervised Learning:**

| Type | Use Cases | Algorithms |
|---|---|---|
| Clustering | Customer Segmentation, Biology, City planning, Target marketing | K-Means, Hierarchical Clustering, DBSCAN, Mean-Shift |
| Association/Dimensionality Reduction | Market Basket Analysis, Text Mining, Face recognition, Big data visualization, Image recognition | Apriori, Eclat, PCA, t-SNE, UMAP |

#### **3. Semi-Supervised Learning**
`Combination of Supervised and Unsupervised Learning`
> Semi-supervised learning is the type of machine learning in which machines are trained on a small amount of labelled data and a large amount of unlabelled data. It is a combination of supervised and unsupervised learning.
- Combination of both
- Small amount of labeled data
- Large amount of unlabeled data
- More accurate than supervised learning
- Less expensive than unsupervised learning

**How Semi-Supervised Learning Works?**



#### **Algorithms used for Semi-Supervised Learning:**
Algorithms for Semi Supervised ML:
- Generative Adversarial networks
- Audio and video manipulation
- Self-trained Naive Bayes classifier 
- Natural language processing

#### **Use-cases of Semi-Supervised Learning:**
- Image recognition
- Speech recognition
- Text categorization
- Sentiment analysis
- Fraud detection

#### **4. Reinforcement Learning**
`Reward based learning`
> Reinforcement learning is the type of machine learning in which machines are trained to make a sequence of decisions. The agent learns to achieve a goal in an uncertain, potentially complex environment. In reinforcement learning, an agent learns by interacting with its environment to achieve a goal.
- Learn from experience / mistakes
- Learn from the consequences of actions
- Hit and trial learning
- Reward and punishment 
- Prediction based on reward and punishment
- Depends on feedback

**How Reinforcement Learning Works?**



#### **Use-Cases of Reinforcement Learning:** 
Mainly used in: 
- Self-driving vehicles
- Gaming
- 3D animations
- Web scrappers 

#### **Algorithms used for Reinforcement Learning:**

Model-Free Reinforcement Learning
- Policy optimizations
- Q-Learning
Model-based Reinforcement Learning
- Learn the Model
- Given the Model
